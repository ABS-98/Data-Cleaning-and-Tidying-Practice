# Introduction:
For this project we're gonna clean and tidy a dataset provided by Stack Overflow, we're gonna practice a few techniques for handling missing data as well as do some basic analysis and visualization.

# Dataset Acknowledgements:
The dataset provided is the result of a series of surveys hosted by Stack Overflow to understand their developer community. For the purposes of this project, the dataset has been slightly modified for the years 2018-2020.

### Goals:
* Prepare and Clean the Data.
* Fill in the missing Data.
* Data Analysis and Visualization.

#### Importing Python Modules:

```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import seaborn as sns
import warnings
warnings.simplefilter(action='ignore', category=FutureWarning)
from pandas.errors import SettingWithCopyWarning
warnings.simplefilter(action='ignore', category=(SettingWithCopyWarning))
```

#### Loading and Inspecting the Data:

```python
df = pd.read_csv('developer_dataset.csv', low_memory=False)
print(df.columns)
```

```python
df.head()
```

*Checking the amount of missing data per column, since this is a survey each question is optional so we are expecting a lot of missing data, the Pandas `.count()` function shows the number of rows in each column which translates to the amount of missing data in each columns.*

```python
df.count()
```

Now let's run a basic summary statistics on the dataset, this will help us understand things like:
* Average Values
* Max and Min Values
* Outliers

```python
df.describe(include = 'all')
```

*The `.count()` function shows the amount of missing data per column, but we can't really compare the numbers as they are, so let's turn them into percentages for easier comparison.*

```python
maxRows = df['RespondentID'].count()
print('% Missing Data:')
print((1 - df.count() / maxRows) * 100)
```

*The columns `(NEWJobHunt, NEWJobHuntResearch, NEWLearn)` have more than 60% missing data, there's no way we can fill these data accurately, so it's safe to drop these columns entirely.*

```python
df.drop(['NEWJobHunt','NEWJobHuntResearch','NEWLearn'],
    axis=1,
    inplace=True)
```

### Analyzing developers by country:

```python
respondent_country = df[['RespondentID','Country']].groupby('Country').count().reset_index()
respondent_country
```

```python
missingData = df[['Employment','DevType']].isnull().groupby(df['Country']).sum().reset_index()
```

```python
plt.figure(figsize = (15,5))
ax1 = plt.subplot(1,2,1)
sns.barplot(data=missingData, x="Country", y="Employment")
ax1.set_xticklabels(missingData.Country, rotation = 30)
ax2 = plt.subplot(1,2,2)
sns.barplot(data=respondent_country, x="Country", y="RespondentID")
ax2.set_xticklabels(respondent_country.Country, rotation = 30)
plt.show()
plt.clf()
```

```python
plt.figure(figsize = (15,5))
ax1 = plt.subplot(1,2,1)
sns.barplot(data=missingData, x="Country", y="DevType")
ax1.set_xticklabels(missingData.Country, rotation = 30)
ax2 = plt.subplot(1,2,2)
sns.barplot(data=respondent_country, x="Country", y="RespondentID")
ax2.set_xticklabels(respondent_country.Country, rotation = 30)
plt.show()
plt.clf()
```

```python
df.dropna(subset = ['Employment','DevType'], inplace = True, how = 'any')
df.count()
```

```python
sns.catplot(x="Country", col="Employment", data=df, kind="count", height=6, aspect=1.5)
plt.show()
plt.clf()
```

```python
devdf = df[['Country','DevType']]
display(devdf.head())
devdf.loc[devdf['DevType'].str.contains('back-end'), 'BackEnd'] = True
devdf.loc[devdf['DevType'].str.contains('front-end'), 'FrontEnd'] = True
devdf.loc[devdf['DevType'].str.contains('full-stack'), 'FullStack'] = True
devdf.loc[devdf['DevType'].str.contains('mobile'), 'Mobile'] = True
devdf.loc[devdf['DevType'].str.contains('administrator'), 'Admin'] = True

devdf = devdf.melt(id_vars=['Country'], value_vars=['BackEnd','FrontEnd','FullStack','Mobile','Admin'], var_name='DevCat',
                   value_name='DevFlag')
devdf.dropna(how='any', inplace=True)
devdf.head()
```

```python
sns.catplot(x="Country", col="DevCat", data=devdf, kind="count", height=6, aspect=1.5)
plt.show()
plt.clf()
```

```python
missingUndergrad = df['UndergradMajor'].isnull().groupby(df['Year']).sum().reset_index()

sns.catplot(x="Year", y="UndergradMajor", data=missingUndergrad, kind="bar", height=4, aspect=1)
plt.show()
plt.clf()
```

```python
df = df.sort_values(['RespondentID','Year'])
df['UndergradMajor'].bfill(axis=0, inplace=True)
```

*Key major groups outlined in the Stack Overflow survey: (social science, natural science, computer science, development, another engineering, never declared)*

```python
edudf = df[['Year','UndergradMajor']]
edudf.dropna(how='any', inplace=True)
display(edudf.head())
edudf.loc[edudf['UndergradMajor'].str.contains('(?i)social science'), 'SocialScience'] = True
edudf.loc[edudf['UndergradMajor'].str.contains('(?i)natural science'), 'NaturalScience'] = True
edudf.loc[edudf['UndergradMajor'].str.contains('(?i)computer science'), 'ComSci'] = True
edudf.loc[edudf['UndergradMajor'].str.contains('(?i)development'), 'ComSci'] = True
edudf.loc[edudf['UndergradMajor'].str.contains('(?i)another engineering'), 'OtherEng'] = True
edudf.loc[edudf['UndergradMajor'].str.contains('(?i)never declared'), 'NoMajor'] = True

edudf = edudf.melt(id_vars=['Year'], value_vars=['SocialScience','NaturalScience','ComSci','OtherEng','NoMajor'], var_name='EduCat',
                   value_name='EduFlag')
edudf.dropna(how='any', inplace=True)
edudf = edudf.groupby(['Year','EduCat']).count().reset_index()
edudf.head()
```

```python
sns.catplot(x="Year", y='EduFlag', col="EduCat", data=edudf, kind="bar", height=6, aspect=1.5)
plt.show()
plt.clf()
```

```python
compFields = df[['Year','YearsCodePro','ConvertedComp']]

plt.figure(figsize = (10,5))
plt.subplot(1,2,1)
sns.boxplot(x="Year", y="YearsCodePro", data=compFields)
plt.subplot(1,2,2)
sns.boxplot(x="Year", y="ConvertedComp", data=compFields)
plt.subplots_adjust(wspace = 0.35)
plt.show()
plt.clf()
```

```python
from sklearn.experimental import enable_iterative_imputer
from sklearn.impute import IterativeImputer
from sklearn.model_selection import train_test_split

imputedf = df[['YearsCodePro','ConvertedComp']]
display(imputedf)
traindf, testdf = train_test_split(imputedf, train_size=0.1)
imp = IterativeImputer(max_iter=20, random_state=0)
imp.fit(imputedf)
compdf = pd.DataFrame(np.round(imp.transform(imputedf),0), columns=['YearsCodePro','ConvertedComp'])
compdf
```

```python
compPlotdf = compdf.loc[compdf['ConvertedComp'] <= 150000]
compPlotdf['CodeYearBins'] = pd.qcut(compPlotdf['YearsCodePro'], q=5)

sns.boxplot(x="CodeYearBins", y="ConvertedComp", data=compPlotdf)
plt.show()
plt.clf()
```
